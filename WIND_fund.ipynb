## Processing


```python
# read price
file_path = "./data/590data/2010-2021_592_fq/"
df_list = []
for file in os.listdir(file_path):
    df = pd.read_csv(file_path + file, index_col=0)
    df["tic"] = file[:-4]
    df_list.append(df)
price_df = pd.concat(df_list).reset_index(drop=True)
price_df = price_df.rename(
    columns=dict(zip(price_df.columns, price_df.columns.map(lambda x: x.lower())))
)
```

```python
fund_df = pd.read_pickle("./data/new_daily_fund.pkl")

target_col = [
    "TICKER",
    "TRADE_DT",
    "VAL_MV",
    "DQ_MV",
    "VAL_PB",
    "VAL_PE_TTM",
    "S_VAL_PCF_OCF",
    "S_VAL_PCF_OCFTTM",
    "S_VAL_PCF_NCF",
    "S_VAL_PCF_NCFTTM",
    "S_VAL_PS",
    "VAL_PS_TTM",
    "TURN"
]

rename_col = [
    "tic",
    "date",
    "MV",
    "C_MV",
    "PB",
    "PE_TMM",
    "PCF_OCF",
    "PCF_OCFTTM",
    "PCF_NCF",
    "PCF_NCFTTM",
    "PS",
    "PS_TTM",
    "TURN"
]

rename_dict = dict(zip(target_col, rename_col))

fund_df = fund_df[target_col]
fund_df = fund_df.rename(columns=rename_dict)
fund_df = fund_df[fund_df.date > "2010-01-01"]
fund_df.head()
```

```python
len(fund_df.tic.unique())
```

```python
dataset_df = pd.merge(price_df, fund_df, how="left", on=['tic', 'date'])
dataset_df.head()
```

```python
import numpy as np

dataset_df = dataset_df.fillna(1e-3)
```

```python
dataset_df.isna().sum()
```

```python
dataset_df.to_pickle("price_with_daily_fund.pkl")
```

```python
total_missing = 0
for tic in dataset_df.tic.unique():
    tic_df = dataset_df[dataset_df.tic == tic]
    l = len(tic_df)
    all_nan_num = (tic_df.isna().sum() == l).sum()
    if all_nan_num > 0:
        print(tic, all_nan_num)
        total_missing += 1
        
print(total_missing)
    
```
